# Datasets

```python
%matplotlib inline
import numpy as np
from plotnine import *
import pandas
```

## Two twodimensional Gaussian distributions

```python
def create_two_2d_gaussians(size_true, size_false=None):
    if size_false is None:
        size_false = size_true
    cov = [[1, 0], [0, 1]]
    true_sample = np.random.multivariate_normal(mean = [0,-2], cov=cov, size = size_true)
    false_sample = np.random.multivariate_normal(mean = [0,2], cov=cov, size = size_false)
    x = np.append(true_sample[:,0], false_sample[:,0])
    y = np.append(true_sample[:,1], false_sample[:,1])
    label = np.append(np.repeat([True], repeats = size_true), np.repeat([False], repeats = size_false))
    return pandas.DataFrame({'x':x,'y':y,'label':label})
```

```python
two_2dgaussians = create_two_2d_gaussians(200)
ggplot(two_2dgaussians, aes(x='x', y='y', fill='label', shape='label'))+geom_point(size=3)
```

```python
two_2dgaussians.to_pickle('data/two_2dgaussians.pkl')
```

## Three twodimensional Gaussian distributions

```python
size = 200
cov = [[1, 0], [0, 1]]
true_sample1 = np.random.multivariate_normal(mean = [0,-2], cov=cov, size = round(size/2))
true_sample2 = np.random.multivariate_normal(mean = [0,9], cov=cov, size = round(size/2))
true_sample = np.append(true_sample1, true_sample2, axis=0)
false_sample = np.random.multivariate_normal(mean = [0,2], cov=cov, size = size)
x = np.append(true_sample[:,0], false_sample[:,0])
y = np.append(true_sample[:,1], false_sample[:,1])
label = np.repeat([True, False], repeats = size)
three_2dgaussians = pandas.DataFrame({'x':x,'y':y,'label':label})
```

```python
ggplot(three_2dgaussians, aes(x='x', y='y', fill='label', shape='label'))+geom_point(size=3)
```

```python
three_2dgaussians.to_pickle('data/three_2dgaussians.pkl')
```

## More clusters

```python
import random
lots_of_gaussians = None
for _ in range(0,8):
    size = random.randint(15,50)
    cov = [[1, 0], [0, 1]]
    xmean = random.random()*10-5
    ymean = random.random()*10-5
    X = np.random.multivariate_normal(mean = [xmean,ymean], cov=cov, size = size)
    label = np.repeat(random.randint(0,5), repeats = size)
    df = pandas.DataFrame({'x':X[:,0],'y':X[:,1],'label':label})
    if lots_of_gaussians is None:
        lots_of_gaussians = df
    else:
        lots_of_gaussians = lots_of_gaussians.append(df, ignore_index=True)
lots_of_gaussians.label = lots_of_gaussians.label.astype('category')    
ggplot(lots_of_gaussians, aes(x='x', y='y', fill='label'))+geom_point(size=3)
```

```python
lots_of_gaussians.to_pickle('data/lots_of_gaussians.pkl')
```
